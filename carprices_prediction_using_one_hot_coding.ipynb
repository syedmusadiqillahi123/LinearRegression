```python
import pandas as pd
df = pd.read_csv("/content/carprices.csv")
df
```

```python
car_model=pd.get_dummies(df['Car Model'])
car_model
```

```python
merged = pd.concat([df,car_model],axis='columns')
merged
```

```python
final = merged.drop(["Car Model","Mercedez Benz C class"], axis='columns')
final
```

```python
X = final.drop(['Sell Price($)'],axis='columns')
X
```

```python
y= final['Sell Price($)']
y
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

```python
model.fit(X,y)
```

```python
model.score(X,y)
```

```python
model.predict([[69000,6,0,1]])
```
